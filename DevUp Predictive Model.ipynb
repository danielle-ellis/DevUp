```python
# check for missing data

def missing_data(df):
    total = df.isnull().sum().sort_values(ascending=False)
    missing_data = pd.concat([total], axis=1, keys=["total"])
    return missing_data
missing_data(df)
```

```python
# determine distance of each trip using formula measured off ellipsoid and latitude/longitude pairs

from pyproj import Geod

wgs84_geod = Geod(ellps='WGS84')

def Distance(lat1,lon1,lat2,lon2):
  az12,az21,dist = wgs84_geod.inv(lon1,lat1,lon2,lat2)
  return dist
```

```python
# convert latitude and longitude to float

x1 = df["start station latitude"].astype(float)
y1 = df["start station longitude"].astype(float)
x2 = df["end station latitude"].astype(float)
y2 = df["end station longitude"].astype(float)
```

```python
x1 = x1.values
y1 = y1.values
x2 = x2.values
y2 = y2.values
```

```python
distance_m = Distance(x1, y1, x2, y2)
```

```python
distance_mi = distance_m/1609.344
```

```python
distmi_df = pd.DataFrame(distance_mi)
distmi_df.columns = ["trip distance (mi)"]
distmi_df.head()
```

```python
# reset index before adding distance to dataframe

df = df.reset_index()
df = df.drop("index",axis =1)
```

```python
df["trip distance"] = distmi_df
```

```python
df.head()
```

```python
# replace missing birth year with median year

median_age = df["birth year"].median()
median_age
```

```python
df["birth year"].replace(np.nan, median_age, inplace = True)
df.head(1)
```

```python
# drop roundtrips because the distance-time metric will skew predictions

df = df.drop(df.index[(df["trip distance"] == 0)])
```

```python
df["hours"] = df["tripduration"]/360
```

```python
df["speed"] = df["trip distance"]/ df["hours"]
```

```python
round(df.describe(),2)
```

```python
# remove trips greater than 40 miles (a distance chosen to remove excessively long trips)

df = df[df["trip distance"] < 40]
```

```python
# remove trips where speed is greater than 25 mph (the average flat-surface biking speed) and less than .01 mph (to remove data errors where no speed was recorded)

df = df[df["speed"]<25]
df = df[df["speed"]> (df["speed"].mean()-(2*df["speed"].std()))]
```

```python
# remove trips longer than 45 minutes (2700 seconds) 
## "The first 30 minutes of each ride are included in the price of the pass.  If you keep a bike out for more than 30 minutes at a time, it's an extra $4 per additional 15 minutes."
### I am assuming trips longer than 45 minutes are mistakes

df = df.drop(df.index[(df["tripduration"] > 2700)])
```

```python
# change hour to time of day 

# (12am-4am = 0, 4am-8am = 1, 8am-12pm = 2, 12pm-4pm = 3, 4pm-8pm = 4, 8pm-12am = 5)

df_model2["hour"] = df_model2["hour"].replace(to_replace=[0,1,2,3], value = 0)
df_model2["hour"] = df_model2["hour"].replace(to_replace=[4,5,6,7], value = 1)
df_model2["hour"] = df_model2["hour"].replace(to_replace=[8,9,10,11], value = 2)
df_model2["hour"] = df_model2["hour"].replace(to_replace=[12,13,14,15], value = 3)
df_model2["hour"] = df_model2["hour"].replace(to_replace=[16,17,18,19], value = 4)
df_model2["hour"] = df_model2["hour"].replace(to_replace=[20,21,22,23], value = 5)
```

```python
# change month to season (winter = 0, spring = 1, summer = 2, fall = 3)

df_model2["month"] = df_model2["month"].replace(to_replace=[12,1,2], value = 0)
df_model2["month"] = df_model2["month"].replace(to_replace=[3,4,5], value = 1)
df_model2["month"] = df_model2["month"].replace(to_replace=[6,7,8], value = 2)
df_model2["month"] = df_model2["month"].replace(to_replace=[9,10,11], value = 3)
```
